```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path050 = './log/fruitbot/nlev_50_easy/vanilla/trial03/progress.csv'
path100 = './log/fruitbot/nlev_100_easy/vanilla/trial02/progress.csv'
path200 = './log/fruitbot/nlev_200_easy/vanilla/trial01/progress.csv'
path500 = './log/fruitbot/nlev_500_easy/vanilla/trial04/progress.csv'
results050 = pd.read_csv(path50)
results100 = pd.read_csv(path100)
results200 = pd.read_csv(path200)
results500 = pd.read_csv(path500)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(results50['total_steps'], results50['eprewmean'], label="50-train")
plt.plot(results100['total_steps'], results100['eprewmean'], label="100-train")
plt.plot(results200['total_steps'], results200['eprewmean'], label="200-train")
plt.plot(results500['total_steps'], results500['eprewmean'], label="500-train")
plt.legend()
plt.title("Steps vs. Mean Episode Reward")
plt.xlabel("Steps")
plt.ylabel("Mean Episode Reward")
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(results50['total_steps'], results50['eval_eprewmean'], label="50-eval")
plt.plot(results100['total_steps'], results100['eval_eprewmean'], label="100-eval")
plt.plot(results200['total_steps'], results200['eval_eprewmean'], label="200-eval")
plt.plot(results500['total_steps'], results500['eval_eprewmean'], label="500-eval")
plt.legend()
plt.title("Steps vs. Eval Mean Episode Reward")
plt.xlabel("Steps")
plt.ylabel("Eval Mean Episode Reward")
plt.show()
```
